```python
import zipfile
import os

def zip_folder(folder_path,output_zip_path):
  try:
    with zipfile.ZipFile(output_zip_path,'w',zipfile.ZIP_DEFLATED) as zipf:
      for root,dirs,files in os.walk(folder_path):
        for file in files:
          file_path = os.path.join(root,file)
          arc_name = os.path.relpath(file_path,folder_path)# get realtive path
          zipf.write(file_path,arc_name)
    print(f"Folder '{folder_path}' has been successfully zipped to '{output_zip_path}'")
  except FileNotFoundError:
    print(f"Folder '{folder_path}' not found")
  except Exception as e:
    print(f"An error occurred: {e}")
```

```python
folder_to_zip = "barcode_yolo_project1"
output_zip_file = "/content/barcode_yolo_project.zip"
zip_folder(folder_to_zip,output_zip_file)
```
