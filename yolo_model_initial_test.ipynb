```python
import ultralytics
from ultralytics import YOLO
import matplotlib.pyplot as plt
import cv2
```

```python
model=YOLO('trained_yolo_model/weights/best.pt')
```

```python
def predict(image_path_batch : list[str]):
    results=model(image_path_batch)
    plt.figure(figsize=(12,6))
    for i, result in enumerate(results,1):
        print(i)
        img=result.plot()
        img_rgb=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
        plt.subplot(1,len(image_path_batch),i)
        plt.imshow(img_rgb)
        plt.axis('off')
    plt.show()
        

```

```python
batch=['assets/images/cut_product.jpg','assets/images/full_product.webp','assets/images/full_product2.png']
```

```python
predict(batch)
```
